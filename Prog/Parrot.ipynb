Задание D

```python
class Parrot:
    def __init__(self, phrase):
        self.phrase = phrase
    
    def say(self, count=1):
        for _ in range(count):
            print(self.phrase, end=" ")
        print()
    
    def newText(self, new_phrase):
        self.phrase = new_phrase

p = Parrot("Гав!")
p.say()          
p.newText("Мяу!")
p.say(3)          
```

Задание Е

```python
import random

class Parrot:
    def __init__(self, phrase):
        self.phrases = [phrase]
    
    def say(self, count=1):
        for _ in range(count):
            print(random.choice(self.phrases), end=" ")
        print()
    
    def learn(self, new_phrase):
        self.phrases.append(new_phrase)

p = Parrot("Гав!")
p.say()           
p.learn("Мяу!")
p.say()           
p.say(3)          
```
