```python
class LampRow:
    def __init__(self):
        self.__state = "00000000"
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == 8:
            self.__state = value
        else:
            self.__state = "00000000"
    
    def show(self):
        result = ""
        for lamp in self.__state:
            if lamp == "0":
                result += "-"
            else:
                result += "*"
        print(result)


lamps = LampRow()
lamps.show()             
lamps.state = "10101010"
print(lamps.state)       
lamps.show()             
```

```python
class LampRow:
    def __init__(self, n=8):
        self.__n = n
        self.__state = "0" * n
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == self.__n:
            self.__state = value
        else:
            self.__state = "0" * self.__n
    
    def show(self):
        result = ""
        for lamp in self.__state:
            if lamp == "0":
                result += "-"
            else:
                result += "*"
        print(result)


lamps = LampRow(6)
lamps.show()              
lamps.state = "101010"
print(lamps.state)        
lamps.show()              
lamps.state = "10101010" 
print(lamps.state)       
lamps.show()             
```

```python
class LampRow:
    def __init__(self, n=8):
        self.__n = n
        self.__state = "0" * n
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, value):
        if len(value) == self.__n:
            self.__state = value
        else:
            self.__state = "0" * self.__n
    
    def show(self):
        result = ""
        for lamp in self.__state:
            if lamp == "0":
                result += "-"
            elif lamp == "1":
                result += "*"
            elif lamp == "2":
                result += "o"
        print(result)


lamps = LampRow(6)
lamps.show()             
lamps.state = "102102"
print(lamps.state)        
lamps.show()             
lamps.state = "10201010" 
print(lamps.state)        
lamps.show()              
```

```python
class LampRow:
    def __init__(self, n=8):
        self.__n = n
        self.__state_int = 0
    
    @property
    def state(self):
        s = str(self.__state_int)
        return s.zfill(self.__n)
    
    @state.setter
    def state(self, value):
        if len(value) == self.__n:
            if value.isdigit():
                self.__state_int = int(value)
            else:
                self.__state_int = 0
        else:
            self.__state_int = 0
    
    def show(self):
        s = self.state  
        result = ""
        for lamp in s:
            if lamp == "0":
                result += "-"
            elif lamp == "1":
                result += "*"
            elif lamp == "2":
                result += "o"
        print(result)


lamps = LampRow(6)
lamps.show()              
lamps.state = "102102"
print(lamps.state)        
lamps.show()              
lamps.state = "10201010"  
print(lamps.state)        
lamps.show()              
```
