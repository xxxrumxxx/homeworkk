```python
import matplotlib.pyplot as plt 
import numpy as np
import math

def f(x):
    return x - 10 * np.sin(x)

print("Графический метод для уравнения x - 10sin(x) = 0")

x = np.linspace(-10, 10, 400)  
y1 = x                          
y2 = 10 * np.sin(x)             

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='x', linewidth=2)
plt.plot(x, y2, label='10sin(x)', linewidth=2)
plt.grid(True)
plt.legend()
plt.title('Отделение корней уравнения x - 10sin(x) = 0')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-10, 10)
plt.ylim(-12, 12)
plt.show()

print("\nИнтервалы, содержащие корни:")
roots_intervals = []
for i in range(len(x)-1):
    if (y1[i] - y2[i]) * (y1[i+1] - y2[i+1]) < 0:
        interval = (round(x[i], 2), round(x[i+1], 2))
        roots_intervals.append(interval)
        print(f"Корень на [{interval[0]}; {interval[1]}]")

print("\nПриблизительные значения корней (пересечение f(x) с осью X):")
x_fine = np.linspace(-10, 10, 1000)
y_fine = f(x_fine)

zero_crossings = []
for i in range(len(x_fine)-1):
    if y_fine[i] * y_fine[i+1] < 0:
        x_root = x_fine[i] - y_fine[i] * (x_fine[i+1] - x_fine[i]) / (y_fine[i+1] - y_fine[i])
        zero_crossings.append(round(x_root, 4))

unique_roots = []
for root in zero_crossings:
    if not unique_roots or abs(root - unique_roots[-1]) > 0.1:
        unique_roots.append(root)

print("Найденные корни:", unique_roots)

plt.figure(figsize=(10, 6))
plt.plot(x_fine, y_fine, 'b-', label='f(x) = x - 10sin(x)', linewidth=2)
plt.axhline(0, color='black', linestyle='-', linewidth=1)
plt.grid(True)

for root in unique_roots:
    plt.plot(root, 0, 'ro', markersize=8)
    plt.annotate(f'{root:.3f}', (root, 0.5), fontsize=9, ha='center')

plt.title('Функция f(x) = x - 10sin(x) и её корни')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim(-10, 10)
plt.legend()
plt.show()

print(f"\nВсего найдено корней: {len(unique_roots)}")
```

```python
import math

def f(x):
    return x - 10 * math.sin(x)

def bisection_method(a, b, eps):
    """
    Метод половинного деления для поиска корня уравнения f(x) = 0
    """
    if f(a) * f(b) >= 0:
        print("Ошибка: функция не меняет знак на концах интервала!")
        return None

    print(f"{'№':<3} {'a':<10} {'b':<10} {'c':<10} {'f(c)':<12} {'|b-a|':<10}")
    print("-" * 60)

    iteration = 0
    while (b - a) / 2 > eps:
        iteration += 1
        c = (a + b) / 2
        fc = f(c)
        width = b - a

        print(f"{iteration:<3} {a:<10.6f} {b:<10.6f} {c:<10.6f} {fc:<12.6f} {width:<10.6f}")

        if fc == 0:
            break
        elif f(a) * fc < 0:
            b = c
        else:
            a = c

    root = (a + b) / 2
    print("-" * 60)
    print(f"\nПриближенное значение корня: {root:.6f}")
    print(f"Точность достигнута: |b - a|/2 = {(b - a) / 2:.6f} < {eps}")
    print(f"Значение функции в корне: f({root:.6f}) = {f(root):.6e}")
    print(f"Количество итераций: {iteration}")
    return root

# Параметры
a = 2.0
b = 3.0
eps = 1e-3

print("Решение уравнения: x - 10·sin(x) = 0")
print(f"Начальный интервал: [{a}, {b}]")
print(f"Требуемая точность: {eps}\n")

bisection_method(a, b, eps)
```
