```python
import math

# Метод простых итераций с правильным преобразованием
# Для уравнения x - 10sin(x) = 0 используем: x = arcsin(x/10) + 2π

def simple_iteration_method(x0, eps=0.001, max_iter=100):
    x = x0
    iteration = 1
    
    print("Решение уравнения x - 10sin(x) = 0 методом простых итераций")
    print("-" * 60)
    print(f"{'Итерация':^10} | {'x':^15} | {'φ(x)':^15} | {'|x - φ(x)|':^15}")
    print("-" * 60)
    
    while iteration <= max_iter:
        # Итерационная функция: φ(x) = arcsin(x/10) + 2π
        # Проверяем, что |x/10| <= 1 для арксинуса
        if abs(x/10) > 1:
            print(f"Ошибка: |x/10| > 1, арксинус не определен")
            return None
            
        x_new = math.asin(x/10) + 2*math.pi
        diff = abs(x_new - x)
        
        print(f"{iteration:^10} | {x:^15.6f} | {x_new:^15.6f} | {diff:^15.6f}")
        
        if diff < eps:
            print("-" * 60)
            return x_new
            
        x = x_new
        iteration += 1
    
    print("-" * 60)
    print("Достигнуто максимальное число итераций")
    return x

# Находим корень
x_root = simple_iteration_method(2.5, 0.001)

print(f"\nКорень: x = {x_root:.6f}")
print(f"Проверка: {x_root:.6f} - 10sin({x_root:.6f}) = {x_root - 10*math.sin(x_root):.6f}")
```
