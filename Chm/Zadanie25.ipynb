Задание 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем 4 функции для 4-х случаев
def f1(x): return x**2 - 1       # f' > 0, f'' > 0 на [1, 2]
def f2(x): return np.log(x + 1)  # f' > 0, f'' < 0 на [0.1, 2]
def f3(x): return 1/x - 0.5      # f' < 0, f'' > 0 на [1, 4]
def f4(x): return 4 - x**2       # f' < 0, f'' < 0 на [0, 2]

funcs = [f1, f2, f3, f4]
titles = [
    "Случай 1: f' > 0, f'' > 0 (Возрастает, вогнута)",
    "Случай 2: f' > 0, f'' < 0 (Возрастает, выпукла)",
    "Случай 3: f' < 0, f'' > 0 (Убывает, вогнута)",
    "Случай 4: f' < 0, f'' < 0 (Убывает, выпукла)"
]

x_ranges = [(0.5, 2), (0.1, 3), (0.5, 4), (0, 2.5)]

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()

for i in range(4):
    a, b = x_ranges[i]
    x = np.linspace(a, b, 100)
    y = funcs[i](x)
    
    axs[i].plot(x, y, 'b-', lw=2)
    axs[i].axhline(0, color='black', lw=1)
    
    # Рисуем хорду (линия между концами отрезка)
    axs[i].plot([a, b], [funcs[i](a), funcs[i](b)], 'g--', label='Хорда')
    
    # Схематичная касательная (в одной из точек)
    axs[i].set_title(titles[i])
    axs[i].grid(True, alpha=0.3)
    axs[i].legend()

plt.tight_layout()
plt.show()
```

```python
def f(x):
    return x**3 + 3*x**2 - 24*x + 1

def df(x):
    return 3*x**2 + 6*x - 24

def trace_iterations(a, b, tol):
    x_cas = a  # точка для касательной (где f*f'' > 0)
    x_hor = b  # точка для хорды
    
    print(f"{'Итер.':<7} | {'x_касательная':<15} | {'x_хорда':<15} | {'Разность'}")
    print("-" * 60)
    
    i = 0
    print(f"{i:<7} | {x_cas:<15.7f} | {x_hor:<15.7f} | {abs(x_cas - x_hor):.7f}")
    
    while abs(x_cas - x_hor) > tol:
        i += 1
        # Формула касательных
        x_cas = x_cas - f(x_cas) / df(x_cas)
        # Формула хорд (используем уже обновленное x_cas для точности)
        x_hor = x_hor - (f(x_hor) * (x_hor - x_cas)) / (f(x_hor) - f(x_cas))
        
        print(f"{i:<7} | {x_cas:<15.7f} | {x_hor:<15.7f} | {abs(x_cas - x_hor):.7f}")

trace_iterations(0, 1, 0.000001)
```
