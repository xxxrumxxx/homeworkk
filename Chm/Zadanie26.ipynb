Задание 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import bisect

# Определяем функцию (lg x это десятичный логарифм log10)
def f1(x):
    return np.log10(x) - np.cos(x)

# 1. Графическое отделение корней
x_vals = np.linspace(0.1, 10, 400)
plt.plot(x_vals, f1(x_vals))
plt.axhline(0, color='red', linestyle='--')
plt.title("График f(x) = lg(x) - cos(x)")
plt.grid(True)
plt.show()

# Видно, что корень лежит примерно между 1 и 2.
# 2. Метод половинного деления (bisect)
root1 = bisect(f1, 1, 2, xtol=1e-4)
print(f"Задача 1: Корень ≈ {root1:.4f}")
```

Задание 2

```python
from scipy.optimize import newton

def f2(x):
    return x**2 - 3.2*x - 1

# 1. Графическое отделение
x_vals = np.linspace(-2, 5, 400)
plt.plot(x_vals, f2(x_vals))
plt.axhline(0, color='red')
plt.grid(True)
plt.show()

# Видим два корня: около -0.3 и около 3.5. 
# Найдем один из них (например, положительный):
root2 = newton(f2, 3.5, tol=1e-4)
print(f"Задача 2: Действительный корень ≈ {root2:.4f}")
```

Задание 3

```python
def f3(x):
    return 4*x - 3*np.log(x) - 4

# 1. Графическое отделение
x_vals = np.linspace(0.1, 3, 400)
plt.plot(x_vals, f3(x_vals))
plt.axhline(0, color='red')
plt.title("График f(x) = 4x - 3ln(x) - 4")
plt.grid(True)
plt.show()

# Корень есть в районе 0.5 и в районе 1.5. Ищем наибольший (около 1.5-2.0).
# 2. Метод простых итераций
def phi(x):
    return (3 * np.log(x) + 4) / 4

def simple_iteration(phi_func, x0, tol):
    x_prev = x0
    while True:
        x_next = phi_func(x_prev)
        if abs(x_next - x_prev) < tol:
            return x_next
        x_prev = x_next

root3 = simple_iteration(phi, 1.5, 1e-4)
print(f"Задача 3: Наибольший корень ≈ {root3:.4f}")
```
