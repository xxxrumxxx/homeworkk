Задание 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Определим функцию f(x) так, чтобы -1 < f'(x) < 0
# Например, f(x) = cos(x) на определенном интервале
def f(x):
    return 0.5 * np.cos(x) + 0.5

x_range = np.linspace(0, 1.5, 100)
x_val = 0.2  # Начальное приближение

plt.figure(figsize=(8, 6))
plt.plot(x_range, f(x_range), label='y = f(x)')
plt.plot(x_range, x_range, '--', label='y = x')

# Рисуем шаги итерации
print("Итерации метода простой итерации:")
print("-" * 40)
print(f"{'Итерация':<10} {'x_n':<15} {'f(x_n)':<15} {'|x_n - x_(n-1)|':<15}")
print("-" * 40)

for i in range(6):
    x_prev = x_val
    y_val = f(x_val)
    
    # Вертикальная линия к функции
    plt.plot([x_val, x_val], [x_val, y_val], 'r')
    # Горизонтальная линия к y=x
    plt.plot([x_val, y_val], [y_val, y_val], 'r')
    
    # Вывод информации об итерации
    if i == 0:
        diff = abs(y_val - x_val)
    else:
        diff = abs(y_val - x_prev)
    print(f"{i+1:<10} {x_val:<15.8f} {y_val:<15.8f} {diff:<15.8f}")
    
    x_val = y_val

print("-" * 40)
print(f"Приближённое решение: x = {x_val:.8f}")

plt.title("метод простой итерации")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from math import *

def phi(x):
    return 1 / sin(x)

def solve_iteration(x0, eps):
    print(f"{'Итерация':<10} | {'x_n':<12} | {'Разность':<12}")
    print("-" * 40)
    
    x_prev = x0
    iteration = 0
    
    while True:
        iteration += 1
        x_next = phi(x_prev)
        diff = abs(x_next - x_prev)
        
        print(f"{iteration:<10} | {x_next:<12.6f} | {diff:<12.6e}")
        
        if diff < eps:
            break
        x_prev = x_next
        
    return x_next

# Параметры задачи
x_start = 1.1
accuracy = 1e-5

root = solve_iteration(x_start, accuracy)
print(f"\nУточненный корень: {root:.6f}")
```
