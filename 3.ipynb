```python
import math

x = 2.5  
eps = 0.001

print("Решение уравнения x - 10*sin(x) = 0 методом простой итерации")
print(f"Начальное приближение: {x}")
print("-" * 50)

iteration = 0
while True:
    x_new = 10 * math.sin(x)
    
    iteration += 1
    print(f"Итерация {iteration}: x = {x:.6f}, x_new = {x_new:.6f}, |x_new - x| = {abs(x_new - x):.6f}")
    
    if abs(x_new - x) < eps:
        break
    x = x_new

print("-" * 50)
print(f"Найденный корень: {x_new:.6f}")
print(f"Проверка: {x_new:.6f} - 10*sin({x_new:.6f}) = {x_new - 10*math.sin(x_new):.6f}")
```
