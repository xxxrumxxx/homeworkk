```python
import math

def f(x):
    return x - 10 * math.sin(x)

def newton_method(x0, eps=0.001, max_iter=100):
    """
    Метод Ньютона (касательных) для надежного нахождения корня
    """
    x = x0
    print("Метод Ньютона:")
    print("-" * 60)
    print(f"{'Итерация':^10} | {'x':^15} | {'f(x)':^15} | {'f\'(x)':^15} | {'Δx':^10}")
    print("-" * 60)
    
    for i in range(max_iter):
        fx = f(x)
        dfx = 1 - 10 * math.cos(x)  # производная f'(x) = 1 - 10cos(x)
        
        if abs(dfx) < 1e-10:  # избегаем деления на ноль
            break
            
        dx = fx / dfx
        x_new = x - dx
        
        print(f"{i+1:^10} | {x:^15.6f} | {fx:^15.6f} | {dfx:^15.6f} | {dx:^10.6f}")
        
        if abs(dx) < eps:
            x = x_new
            break
            
        x = x_new
    
    print("-" * 60)
    return x

# Находим корень около 2.5
x_root = newton_method(2.5, 0.001)

print(f"\nНайденный корень: x = {x_root:.6f}")
print(f"Проверка: f({x_root:.6f}) = {x_root:.6f} - 10sin({x_root:.6f}) = {f(x_root):.6f}")
```
